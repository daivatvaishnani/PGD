```python
import pandas as pd
import numpy as np
```

```python
censusData = pd.read_csv('EDA_census.csv')
censusData.head(10)
```

```python
censusData.dtypes
```

# Calculating % Females within age group 20-24 who are illiterate in India

```python
censusData['Age-group'].value_counts()
```

```python
censusDataForAgeGroup = censusData[(censusData['Area Name'] == 'INDIA') & (censusData['Age-group'] == '20-24')]
censusDataForAgeGroup
```

```python
totalFemales = censusDataForAgeGroup.loc[censusData['TRU'] == 'Total', 'Total Females']
totalFemales
```

```python
totalIlliterateFemales = censusDataForAgeGroup.loc[censusData['TRU'] == 'Total', 'Illiterate\t\tFemales']
totalIlliterateFemales
```

```python
percentageIlliteracyInFemales = round((totalIlliterateFemales / totalFemales) * 100)
percentageIlliteracyInFemales
```

# Comparing literacy rates in all age groups

```python
censusDataForIndia = censusData[(censusData['Area Name'] == 'INDIA') & (censusData['TRU'] == 'Total')]
censusDataForIndia
```

```python
censusDataForIndiaGroupWise = censusDataForIndia.loc[:,['Age-group', 'Total Persons', 'Illiterate\t\tPersons', 'Literate\t\tPersons']]
censusDataForIndiaGroupWise['LiteracyRate'] = censusDataForIndiaGroupWise['Literate\t\tPersons'] / censusDataForIndiaGroupWise['Total Persons']
censusDataForIndiaGroupWise
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(20,20))
plt.bar(censusDataForIndiaGroupWise['Age-group'], censusDataForIndiaGroupWise['LiteracyRate'])
```

# Comparing states on female literacy rates

```python
censusDataForFemales = censusData.loc[(censusData['Age-group'] == 'All ages') & (censusData['TRU'] == 'Total'), ['Area Name','Total Females', 'Illiterate\t\tFemales', 'Literate\t\tFemales']]
censusDataForFemales['LiteracyRate'] = censusDataForFemales['Literate\t\tFemales'] / censusDataForFemales['Total Females']
censusDataForFemales
```

```python
censusDataForFemales.sort_values(by=['LiteracyRate'],ascending=False).head()
```

```python
import numpy as np
```

```python
def isMainCategory(category):
    subs = category.split("|")
    tot = np.sum([True for x in subs if x in categories])
    return True if tot > 0 else False
```

```python
def getMainSector(primary_sector):
    entry = mapping[mapping.category_list == primary_sector]
    return [x for x in entry.columns if (entry.loc[:,x] == 1).values[0]][0]
```

```python
main_sectors = mapping.category_list.values
```

```python
'Active Lifestyle' in main_sectors
```

```python
top3['primary_sector'] = top3.category_list.apply(lambda x : x.split('|')[0])
```

```python
top3MC = top3[top3.primary_sector.isin(main_sectors)]
```

```python
top3MC
```

```python
top3MC['main_sector'] = top3MC.primary_sector.apply(lambda x : getMainSector(x))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
data = pd.read_csv('train.csv')
data
```

```python
# finding missing values and what is the dtype of the respective columns

cols_with_nulls = data.loc[:,data.isnull().sum() > 0].columns
pd.concat([data[cols_with_nulls].isnull().sum() / len(data) * 100, data[cols_with_nulls].dtypes],
          axis = 1, keys=['missing%', 'dtype'])
```

```python
# we will maintain a list of columns that we drop
cols_dropped = []
```

```python
cols_to_drop = ['Alley', 'PoolQC', 'Fence', 'MiscFeature']
data.drop(columns=cols_to_drop, inplace=True)
cols_dropped += cols_to_drop
```

```python
cols_with_nulls = data.loc[:,data.isnull().sum() > 0].columns
pd.concat([data[cols_with_nulls].isnull().sum() / len(data) * 100, data[cols_with_nulls].dtypes],
          axis = 1, keys=['missing%', 'dtype'])
```

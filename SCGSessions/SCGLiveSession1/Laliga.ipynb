```python
import pandas as pd
```

```python
laliga = pd.read_csv('Laliga-2.csv')
```

```python
laliga_cols = laliga.iloc[0]
laliga = laliga[1:]
laliga.columns = laliga_cols
laliga
```

```python
laliga
```

```python
laliga2 = laliga
```

```python
laliga2['Champion'] = laliga2['Champion'].apply(lambda x: 0 if x == '-' else int(x))
laliga2['Runner-up'] = laliga2['Runner-up'].apply(lambda x: 0 if x == '-' else int(x))
laliga2['Third'] = laliga2['Third'].apply(lambda x: 0 if x == '-' else int(x))
laliga2['Fourth'] = laliga2['Fourth'].apply(lambda x: 0 if x == '-' else int(x))
laliga2['Fifth'] = laliga2['Fifth'].apply(lambda x: 0 if x == '-' else int(x))
laliga2['Sixth'] = laliga2['Sixth'].apply(lambda x: 0 if x == '-' else int(x))
```

```python
laliga2
```

```python
laliga3 = laliga2.iloc[:60]
```

```python
laliga3
```

```python
laliga3.dtypes
```

```python
laliga3['Seasons'] = laliga3['Seasons'].astype(int)
laliga3['Points'] = laliga3['Points'].astype(int)
laliga3['GamesPlayed'] = laliga3['GamesPlayed'].astype(int)
laliga3['GamesWon'] = laliga3['GamesWon'].astype(int)
laliga3['GamesLost'] = laliga3['GamesLost'].astype(int)
laliga3['GamesDrawn'] = laliga3['GamesDrawn'].astype(int)
laliga3['GoalsFor'] = laliga3['GoalsFor'].astype(int)
laliga3['GoalsAgainst'] = laliga3['GoalsAgainst'].astype(int)
```

```python
laliga3.dtypes
```

```python
laliga3.to_csv('laliga3.csv', index=False)
```
